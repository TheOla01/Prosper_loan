### Preliminary Rangling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Prosper Loan.csv')
```

```python
df['BorrowerAPR'] = (df['BorrowerAPR'].divide(df['Total'], axis=0)*100).round(2).astype(str) + ' %'
```

```python
df= df.drop(columns=['ListingKey','ListingNumber','ListingCreationDate','PercentFunded', 'Recommendations','Total'])
```

## Univariate Exploration

### Categorical Variable

```python
#Bar Chart
sns.countplot(data=df, x='creditgrade', color= 'blue')
plt.title('Distribution of credit grades');
```

## Observation
    The above bar chat shows that Grade C has the highest occurence more than the rest grades.

```python
#Pie Chart
plt.pie(df.term.value_counts(),
       labels=df.term.value_counts().index,
       wedgeprops= {'width':0.5});
plt.title('Distribution of Term');
```

## Observation 
    The above Pie chat shows that 36 Term has the highest occurrence which means majority of the loans falls in the 36 term

# Quantitative Variable

```python
#histogram
plt.hist(data=df, x='borrowerrate', bins= 10);
plt.xlabel('Borrower Rate')
plt.ylabel('count')
plt.title('Distribution of Borrower Rate');
```

## Observation 
    The histogram above shows that majority of the borrowing rate falls within the rate of 0.1 and 0.2.

# Bivariate Exploration

### Two Categorical variable 

```python
#Cluster BarChat
sns.barplot(data=df, x= 'term', y= 'loanstatus', hue= 'term')
plt.title('Distribution of Loan Status by Term');
```

## Observation
    The clustered bar chat shows the relationship between loan terms and loan status, from the chat all cancelled loans are 36 term which means there should be more attention should be placed on the 36 term loans or if possible it should be eradicated.

### Two Quantitative Variable

```python
#scattered plot
sns.regplot(data=df, x='investmentfromfriendsamount', y='lenderyield')
plt.title('Distribution of Investment from friends by Lender Yield');
```

## Observation
    The above scattered plot shows the relationship between the investment from friends and Lender yield. Friends invest higher amount on loans with low lender yield as inferred from the plot.

### Two Categorical Variable

```python
#Scatter plot
sns.regplot(data=df, x='loanoriginalamount', y='statedmonthlyincome')
plt.title('Distribution of Loan Original Amount by Stated Monthly Income');
```

## Observation
    Above scattered plot is used to show the relatinship betweeb the loan original amount and stated monthly income. The chat shows that the majority of the loan is given to applicants with low monthly income.

### One Categorical Variable and One Quantitative Variable

```python
#Violin chat
sns.violinplot(data=df, x= 'loanoriginalamount', y= 'employmentstatus')
plt.title('Distribution of Loan Original Amount by Employment Status');
```

## Observation
    The above violin chat is used to show the relationship between loan original amount and employment status. This shows that those with the employment status 'employed' and 'full-time' has higher priviledge to get huge amount of loan.

### One Categorical and One Quantitative Variable

```python
#Box plot
sns.boxplot(data=df, x='isborrowerhomeowner', y= 'loanoriginalamount')
plt.title('Distribution of Loan Original Amount by Is Borrower Home Owner');
```

## Observation
    The above box plot is used to show the relationship between loan original amount and is borrower home owner. the plot shows that the loan is mostly given to those who are home owner and home owners have the priviledge to get higher amount of loan.

# Multivariate Exploration


### Two Categorical and One Quantitative Variable

```python
#Point Plot
sns.pointplot(data=df, x='loanoriginalamount', y='employmentstatus', hue='isborrowerhomeowner')
plt.title('Distribution of Loan Original Amount by Employment Status and is borrower home owner');
```

## Observation
    The above point plot is used to show the relationship between loan original amount, employment status and is borrower home owner. The plot shows that home owners that are employed get highest amount of loan.

```python
#Point plot
sns.pointplot(data=df, x='lp_netprincipalloss', y='employmentstatus', hue='isborrowerhomeowner')
plt.plot('Distribution of Lp_net principal loss by employment status and is borrower home owner');
```

## Observation
    The point plot above shows the relationship between lp_net principal loss, employment status and is borrower home owner. The plot shows that the higher amount of net principal loss came from home owner with employment status 'Not available'.
